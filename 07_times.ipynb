```python
deviceid_packages=pd.read_csv(FLAGS.file_path+'deviceid_packages.csv')
def app_count(text):
    app_list=text.split('|')
    return len(app_list)
deviceid_packages_07['app_len']=deviceid_packages['add_id_list'].apply(lambda line:app_count(line))
deviceid_packages_07['every_app_len']=deviceid_packages_07['times_len']/deviceid_packages_07['app_len']
```

```python
deviceid_packages_07.describe()
```

```python
device_id=deviceid_packages_07.ix[:,'device_id']
deviceid_packages_07.drop('device_id', axis=1,inplace = True)

deviceid_packages_07=deviceid_packages_07/deviceid_packages_07.std()
deviceid_packages_07=pd.concat([device_id,deviceid_packages_07],axis=1)
```

```python
deviceid_packages_07=deviceid_packages_07.fillna(0)
```

```python
deviceid_packages_07.describe()
```

```python
columns=['device_id','app_len', 'every_app_len', 'times_len', 'app_time_t2_61', 'app_time_t1_32', 'app_time_t1_33', 'app_time_t2_132', 
             'app_time_t2_94', 'app_time_t1_36', 'app_time_t1_43', 'app_time_t2_223', 'app_time_t2_124', 'app_time_t2_251', 'app_time_t2_158',]
```

```python
deviceid_packages_07.to_csv(FLAGS.file_path+'07_deviceid_packages.csv',columns=columns,index= False)
```
